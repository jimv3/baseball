```python
import baseball
game = baseball.Game()
game.list_teams()
```

```python
players = game.get_players('aTL')
display(players)
```

```python
p = players.loc[players['Player'] == 'Freddie Freeman']
```

```python
import random as rnd
```

```python
num_hits = 0
at_bat_count = 0
num_iterations = 10_000

def is_hit(at_bat):
    hits = ['1B', '2B','3B', 'HR']
    return at_bat in hits

for i in range(num_iterations):
    at_bat = game.at_bat(p)
    if is_hit(at_bat):
        num_hits += 1
    if at_bat != 'BB':
        at_bat_count += 1
        
        
print(f'Random at bat AVG: {num_hits / at_bat_count:.3f}')
print(f'Actual season AVG: {p["AVG"]}')
```
